```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1='2017_election_data.csv'
file2="2022_election_data.csv"
```

```python
df1=pd.read_csv(file1)
df1
```

```python
df2=pd.read_csv(file2)
df2
```

```python
votes_asc_2017=df1.sort_values(by=['Votes'], ascending=False)
votes_asc_2017
```

```python
plt.pie(votes_asc_2017['Votes'], labels=votes_asc_2017['Party'])
plt.show()
```

```python
seats_asc_2017=df1.sort_values(by=['Seats'], ascending=False)
seats_asc_2017
```

```python
plt.pie(seats_asc_2017['Seats'], labels=seats_asc_2017['Party'])
plt.show()
```

```python
votes_asc_2022=df2.sort_values(by=['Votes'], ascending=False)
votes_asc_2022
```

```python
plt.pie(votes_asc_2022['Votes'], labels=votes_asc_2022['Party'])
plt.show()
```

```python
seats_asc_2022=df2.sort_values(by=['Seats'], ascending=False)
seats_asc_2022
```

```python
plt.pie(seats_asc_2022['Seats'], labels=seats_asc_2022['Party'])
plt.show()
```
